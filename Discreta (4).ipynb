```python
%matplotlib inline
import numpy as np
import scipy as sp
import scipy.linalg as la
import networkx as nx
import random
from nxpd import draw
import scipy.integrate as sint
import matplotlib.pyplot as plt
```

```python
print("Grafo RTVE")
G = nx.DiGraph()

#Nombre de los nodos, posicion y color
G.add_node("PSOE",pos=(2,3),color='red')
G.add_node("Cs",pos=(3,3),color='orange')
G.add_node("PP",pos=(4,2),color='blue')
G.add_node("UP",pos=(1,2),color='purple')
G.add_node("VOX",pos=(3,1),color='green')
G.add_node("OT",pos=(2,1),color='black')

#Aristas PSOE
G.add_edge("PSOE","PP", weight=18)
G.add_edge("PSOE","Cs", weight=10)
G.add_edge("PSOE","UP", weight=3)
G.add_edge("PSOE","VOX", weight=6)
G.add_edge("PSOE","OT", weight=1)

#Aristas Cs
G.add_edge("Cs","PSOE", weight=21)
G.add_edge("Cs","UP", weight=7)
G.add_edge("Cs","PP", weight=9)
G.add_edge("Cs","OT", weight=17)

#Aristas PP
G.add_edge("PP","PSOE", weight=21)
G.add_edge("PP","UP", weight=3)
G.add_edge("PP","Cs", weight=1)
G.add_edge("PP","OT", weight=16)

#Aristas UP
G.add_edge("UP","PP", weight=5)
G.add_edge("UP","Cs", weight=7)
G.add_edge("UP","PSOE", weight=11)
G.add_edge("UP","VOX", weight=1)
G.add_edge("UP","OT", weight=4)



pos=nx.get_node_attributes(G,'pos')

blue_nodes=[n for n,d in G.nodes(data=True) if d['color']=='blue']
red_nodes=[n for n,d in G.nodes(data=True) if d['color']=='red']
orange_nodes=[n for n,d in G.nodes(data=True) if d['color']=='orange']
purple_nodes=[n for n,d in G.nodes(data=True) if d['color']=='purple']
green_nodes=[n for n,d in G.nodes(data=True) if d['color']=='green']
black_nodes=[n for n,d in G.nodes(data=True) if d['color']=='black']


#Imprimimos los nodos y su posicion
nx.draw(G,pos,nodelist=blue_nodes,node_color='b')
nx.draw(G,pos,nodelist=red_nodes,node_color='r')
nx.draw(G,pos,nodelist=orange_nodes,node_color='orange')
nx.draw(G,pos,nodelist=purple_nodes,node_color='purple')
nx.draw(G,pos,nodelist=green_nodes,node_color='green')
nx.draw(G,pos,nodelist=black_nodes,node_color='black')


#Imprimimos las aristas
labels=nx.get_edge_attributes(G,'weight')
nx.draw_networkx_edge_labels(G,pos,edge_labels=labels)
a=nx.adjacency_matrix(G)
a.todense()
#nx.to_numpy_matrix(G)
```
